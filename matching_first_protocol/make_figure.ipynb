## 小提琴图

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.ticker as mticker
import seaborn as sns
import pickle


def save_figure(name,metric1, metric2, metric3):

    plt.figure(figsize=(8, 6))
    # 将数据转换为长格式
    df = [ np.array(metric1) ,np.array(metric2), np.array(metric3)]

    # 设置颜色和透明度
    colors = ['#8ECFC9', '#FFBE7A', '#82B0D2']

    # 绘制小提琴图，调整带宽以减少波动bandwidth #  cut=0 表示在0处截断 #inner=None 表示不显示中位数
    ax = sns.violinplot( data=df, labels=["Uni.: 0.8", "M1", "M2"], palette=colors, inner=None,
                        linewidth=1.0, bw=0.2,cut=0)

    ax.set_xticklabels(["Uni.: 0.8", "M1", "M2"])

     # 计算并标注均值
    means = [column.mean() for column in df]
    for i, mean in enumerate(means):
        # 画交叉线表示均值
        # plt.plot([i + 1 - 0.1, i + 1 + 0.1], [mean, mean], color='red', linestyle='-', linewidth=1)  # 水平线
        plt.plot([i - 0.02, i + 0.02], [mean, mean], color='black')
        plt.text(i, mean, f'{mean:.2f}', color='black', ha='center', va='bottom')
        # plt.plot([i + 1, i + 1], [mean - 0.1, mean + 0.1], color='red', linestyle='-', linewidth=1)  # 垂直线
        # # plt.plot([i + 1, i + 1], [mean -0.005, mean +0.005], color='red', linestyle='-', linewidth=1)  # 较短的垂直线
        # plt.text(i + 1.03, mean, f'{mean:.2f}', color='black', ha='center', va='bottom')  # 添加均值标签

    # 例如：设置刻度为从0到100，间隔为10
    # plt.yticks(range(0, 101, 10))

    # 显示图表
    plt.xlabel('Pricing Strategy', fontsize=14)
    plt.ylabel(name, fontsize=14)

    plt.savefig(f'./figure/comparision/The violin plot of {name}.png', dpi= 300)
    plt.close()


# Load data for method 1
with open('./output/2017-05-02/system_metric_daytoday.pkl', 'rb') as file:
    day_to_day = pickle.load(file)
    


# Load data for method 2
with open('./output/2017-05-02/system_metric_method2.pkl', 'rb') as file:
    method2 = pickle.load(file)
    # print(method2)
    # print(method2.keys())
    

# Load data for method 2
with open('./output/2017-05-02/system_metric_uniformed.pkl', 'rb') as file:
    uniformed = pickle.load(file)
    # print(uniformed)
    # print(uniformed.keys())

# Extract metrics for method 2
waitingTime1 = [time / 60 for time in uniformed['waitingTime']]
waitingTime2 = [time / 60 for time in day_to_day['waitingTime']]
waitingTime3 = [time / 60 for time in method2['waitingTime']]
save_figure('Waiting Time for Response (s)', waitingTime1, waitingTime2,waitingTime3)


detour_distance1 = [distance / 1000 * 2 for distance in uniformed['detour_distance']]
detour_distance2 = [distance / 1000 * 2 for distance in day_to_day['detour_distance']]
detour_distance3 = [distance / 1000 * 2 for distance in method2['detour_distance']]


save_figure('Detour time of each order (min)', detour_distance1, detour_distance2,detour_distance3)


shared_distance1 = [distance / 1000 * 2 for distance in uniformed['shared_distance']]
shared_distance2 = [distance / 1000 * 2 for distance in day_to_day['shared_distance']]
shared_distance3 = [distance / 1000 * 2 for distance in method2['shared_distance']]

save_figure('Shared Distance (m)', shared_distance1, shared_distance2,shared_distance3)

platform_income1 = uniformed['platform_income'] 
platform_income2 = day_to_day['platform_income'] 
platform_income3 = method2['platform_income'] 
save_figure('Platform Profit', platform_income1, platform_income2,platform_income3)

saved_ride_distance1 = uniformed['saved_ride_distance'] 
saved_ride_distance2 = day_to_day['saved_ride_distance'] 
saved_ride_distance3 = method2['saved_ride_distance'] 
save_figure('Distance Savings (m)', saved_ride_distance1, saved_ride_distance2,saved_ride_distance3)

pickup_time1 = [time / 60 for time in uniformed['pickup_time']]
pickup_time2 = [time / 60 for time in day_to_day['pickup_time']]
pickup_time3 = [time / 60 for time in method2['pickup_time']]

save_figure('Waiting time of each order (min)', pickup_time1, pickup_time2,pickup_time3)

traveltime1 = [time / 60 for time in uniformed['traveltime']]
traveltime2 = [time / 60 for time in day_to_day['traveltime']]
traveltime3 = [time / 60 for time in method2['traveltime']]

save_figure('Ride time of each order (min)', traveltime1, traveltime2,traveltime3)

discount_ratio1 = uniformed['Discount ratio'] 
discount_ratio2 = day_to_day['Discount ratio']
discount_ratio3 = method2['Discount ratio']
save_figure('Discount ratio', discount_ratio1, discount_ratio2,discount_ratio3)     
```

```python
import pandas as pd
orders = pd.read_csv('../order.csv')
node = pd.read_csv('../node.csv')
print(orders.columns)
orders.head()
```

```python
def returnKey(dic, value):
       for i,j in dic.items():
              if j==value:
                     return i

tmp = orders
i = 0
for index, row in tmp.iterrows():
       # print(index)
       if i%1000 == 0:
              print(i)
       i += 1
       start_distance = {}
       end_distance = {}
       for index1,row1 in node.iterrows():
              start_distance[row1['node_id']] = \
            abs(row["dwv_order_make_haikou_1.starting_lng"]  -   row1['x_coord']) + \
            abs(row['dwv_order_make_haikou_1.starting_lat']  -   row1['y_coord'])
              end_distance[row1['node_id']] = \
            abs(row['dwv_order_make_haikou_1.dest_lng']  -   row1['x_coord']) + \
            abs(row['dwv_order_make_haikou_1.dest_lat']  -   row1['y_coord'])
       # print(start_distance)
       # print(returnKey(start_distance, min(start_distance.values())))
       tmp.loc[index,'O_location'] = returnKey(start_distance, min(start_distance.values()))
       tmp.loc[index,'D_location'] = returnKey(end_distance, min(end_distance.values()))
       
tmp.head()
```

```python
tmp[['dwv_order_make_haikou_1.normal_time','dwv_order_make_haikou_1.arrive_time',
       'dwv_order_make_haikou_1.departure_time','dwv_order_make_haikou_1.start_dest_distance','dwv_order_make_haikou_1.pre_total_fee',]].head()
```

```python
od = pd.read_csv('../OD.csv')
# 使用 rename 方法为列重命名
od = od.rename(columns={
    'origin_id': 'O_location',
    'destination_id': 'D_location',
})
od.head()
```

```python
merged = pd.merge(left= tmp, right= od, on=['O_location','D_location'])
merged.head()
```

```python
predict = pd.read_csv('../predict_result.csv')
predict.head()
```

```python
merged_results = pd.merge(left = merged, right= predict, on=['OD_id'])
merged_results.head()
```

```python
merged_results.to_csv('orders_predict.csv')
```

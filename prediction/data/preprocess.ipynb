```python
import math

def acceptance_probability(discount, k=5, x0=1):
    """
    计算接受价格折扣的概率
    
    Parameters:
    - discount: 价格折扣
    - k: sigmoid 函数的斜率参数
    - x0: sigmoid 函数的中心位置
    
    Returns:
    - 概率在 0 到 1 之间
    """
    exponent = -k * (x0 - discount)
    probability = 1 / (1 + math.exp(exponent))
    return probability

# 测试
for discount in [0.5, 0.6, 0.7, 0.8, 0.9]:
    probability = acceptance_probability(discount)
    print(f"当价格折扣为 {discount} 时，接受概率为: {probability:.2%}")
```
